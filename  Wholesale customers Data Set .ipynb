### Objective
In this notebook the goal is to analyze annual spending products to predict the Channel(i.e. Retail, Hotel). The first part is classification, where three models are built using: grid search, train test split, and cross validation. Second, the data will be clustered.

### Needed Libraries

```python
import numpy as np
import pandas as pd 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.cluster import KMeans
```

### Data Import

```python
df = pd.read_csv('Wholesale customers data.csv')
```

### Data Setup

```python
X = df.iloc[:, 2:]
X.head()
```

```python
y_1 = df.iloc[:, 0:1]
y_1.head()
```

```python
y_2 = df.iloc[:, 1:2]
y_2.head()
```

## Classification

### Classification; GridSearch

```python
parameters= {'n_neighbors': np.arange(1,5)}
```

```python
knn = KNeighborsClassifier()
knn_cv_1 = GridSearchCV(knn, parameters, cv=5)
knn_cv_2 = GridSearchCV(knn, parameters, cv=5)

knn_cv_1.fit(X, y_1)
knn_cv_2.fit(X, y_2)
```

```python
print("Model for Channel best parameters: {}".format(knn_cv_1.best_params_))
print("Model for Channel best score: {}".format(knn_cv_1.best_score_))
```

```python
print("Model for Region best parameters: {}".format(knn_cv_2.best_params_))
print("Model for Region best score: {}".format(knn_cv_2.best_score_))
```

For both models, the best number of neighbors is 1 unit above of the actual value.

### Classification; Train Test Split

```python
# Train test split set up
X_1_train, X_1_test, y_1_train, y_1_test = train_test_split(X, y_1, test_size = 0.3, random_state = 42)
X_2_train, X_2_test, y_2_train, y_2_test = train_test_split(X, y_2, test_size = 0.3, random_state = 42)
```

```python
# Models with actual values
knn_a_1 = KNeighborsClassifier(n_neighbors = 2)
knn_a_2 = KNeighborsClassifier(n_neighbors = 3)

# Models fitting
knn_a_1.fit(X_1_train, y_1_train)
knn_a_2.fit(X_2_train, y_2_train)

# Prediction
y_a_1_pred = knn_a_1.predict(X_1_test)
y_a_2_pred = knn_a_2.predict(X_2_test)

# Models score
y_a_1_score = knn_a_1.score(X_1_test, y_1_test)
y_a_2_score = knn_a_1.score(X_1_test, y_2_test)

print('R squared for Channels: {}'. format(y_a_1_score))
print('R squared for Regions: {}'. format(y_a_2_score))
```

### Classification; Cross Validation

```python
# Models with actual values
knn_a_1_cv = KNeighborsClassifier(n_neighbors = 2)
knn_a_2_cv = KNeighborsClassifier(n_neighbors = 3)

# Models cross validation score
cv_1 = cross_val_score(knn_a_1_cv, X, y_1, cv=5)
cv_2 = cross_val_score(knn_a_2_cv, X, y_2, cv=5)

print('R squared for Channels: {}'. format(cv_1))
print('R squared for Regions: {}'. format(cv_2))
```

## Clustering

```python
# Models with actual values
km_1 = KMeans(n_clusters = 2)
km_2 = KMeans(n_clusters = 3)

# Models fitting
km_1.fit(X)
km_2.fit(X)

# Models prediction
labels_1 = km_1.predict(X)
labels_2 = km_2.predict(X)
```

```python
# Labels modification, since the labels start from 0 by adding 1 makes it easier to compare to the actual labels.
labels_11 = labels_1+1
labels_22 = labels_2+1

# Labels Data Frame
df_l = pd.DataFrame({'Channel_l':labels_11, 'Region_l':labels_22})

# Labels comparison
df_c = pd.crosstab(df['Channel'], df_l['Channel_l'])
df_r = pd.crosstab(df['Region'], df_l['Region_l'])

print(df_c)
print(" \n")
print(df_r)
```

### Discussion

From the GridSearch approach we learned that the suggested number of neighbors is one unit above the actual value, this might be due to to the fact that we the GridSearch parameter is limited to a maximum of 5 neighbors. From train test split section, the Region labels has poor peerformance compared to the others methods. In contrast, cross validation yielded a better performance for the Region labels and for Channel is very similar to the others methods. 

The clustering section is not an accurate model, however, this data can be used to predict shopping behaviour. One approach to determine the number of types of shoppers would be, first used GridSearch to determine the optimal number of neighbors then proceed with that same number as the number of clusters to label shoppers.  

```python
# GridSearchCV parameters
params = {'n_neighbors': np.arange(1, 50)}

# Model
knn_n = KNeighborsClassifier()
knn_n_cv = GridSearchCV(knn_n, params, cv=5)

# Model fitting
knn_n_cv.fit(X, y_1)

print("Model for Channel best parameters: {}".format(knn_n_cv.best_params_))
print("Model for Channel best score: {}".format(knn_n_cv.best_score_))
```

The model with the best performance seems to be 15 neighbors.

```python
# Clustering model
km_sh = KMeans(n_clusters=15)

# Model fitting
km_sh.fit(X)

# Clusters labels
labels_sh = km_sh.predict(X)

print('Types of shoppers: {}'.format(labels_sh))
```
